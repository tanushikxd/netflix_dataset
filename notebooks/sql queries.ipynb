```python
import sqlite3
import pandas as pd

df = pd.read_csv('../data/netflix_cleaned.csv')
conn = sqlite3.connect('../data/netflix.db')
df.to_sql('netflix', conn, if_exists='replace', index=False)

# query = "SELECT type, COUNT(*) as count FROM netflix GROUP BY type"
# pd.read_sql_query(query, conn)

query = """
SELECT type, COUNT(*) AS count
FROM netflix
GROUP BY type
ORDER BY count DESC
"""
pd.read_sql(query, conn)
```

Shows that Netflix has more movies than Tv Shows

```python
query = """ 
SELECT country, COUNT(*) AS count
FROM netflix
WHERE country IS NOT NULL
GROUP BY country
ORDER BY count DESC
LIMIT 10
"""
pd.read_sql(query, conn)
```

List of countries with majority of content available

```python
query = """ 
SELECT rating, COUNT(*) AS count
FROM netflix
GROUP BY rating
ORDER BY count DESC
LIMIT 10
"""
pd.read_sql(query, conn)
```

The most popular ratings TOP 10

```python
query = """ 
SELECT rating, type, COUNT (*) AS count
FROM netflix
GROUP BY rating, type
ORDER BY rating
LIMIT 15
"""
pd.read_sql(query, conn)
```
