```python
#compare the amout of movies and tv series

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('../data/netflix_cleaned.csv')

plt.figure(figsize=(6,4))
sns.countplot(data=df, x='type')
plt.title('Movies vs TV Shows')
plt.show()
```

On the bar Movies vs Tv shows, we can clearly see that movie content is double higher on Netflix platform.

```python
#Movies added by year

year_counts = df['year_added'].value_counts().sort_index()
year_counts.plot(kind='bar', figsize=(12,5))
plt.title('Content added per year')
plt.show()
```

The graph is showing how drastically content on platform increased by years, for example on 2014 it was on its minimum, the max it reached on 2019 and dropped down a bit during peak of COVID.

```python
#top-10

from collections import Counter

genres_series = df['listed_in'].dropna().apply(lambda x: [g.strip() for g in x.split(',')]) # splited genres for the list

all_genres = [g for sublist in genres_series for g in sublist] # attaching all genres at one

genre_counts = Counter(all_genres) #counting frequency

import pandas as pd
import matplotlib.pyplot as plt

genre_df = pd.DataFrame(genre_counts.items(), columns=['Genre', 'Count']).sort_values('Count', ascending=False)

plt.figure(figsize=(10,5))
plt.barh(genre_df['Genre'].head(10), genre_df['Count'].head(10), color='cornflowerblue')
plt.gca().invert_yaxis()
plt.title('Top 15 Genres on Netflix')
plt.xlabel('Number of Titles')
plt.show()

```

As we can see on the plot, the majority of movies are from international segment, but if we lookg at genres, the most popular os Dramas. With a liitle difference, we can see Comedies and Documentaries.
